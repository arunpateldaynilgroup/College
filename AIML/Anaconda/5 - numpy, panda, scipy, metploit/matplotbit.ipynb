```python
# importing matplotlib module 
from matplotlib import pyplot as plt 
```

```python
# x-axis values 
x = [5, 2, 9, 4, 7] 
```

```python
# Y-axis values 
y = [10, 5, 8, 4, 2] 
```

```python
# Function to plot 
plt.plot(x, y) 
```

```python
# function to show the plot 
plt.show()
```

```python
#Histogram
from matplotlib import pyplot as plt 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot histogram 
plt.hist(y) 
  
# Function to show the plot 
plt.show() 
```

```python
#Scatter Plot
# x-axis values 
x = [5, 2, 9, 4, 7] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot scatter 
plt.scatter(x, y) 
  
# function to show the plot 
plt.show() 
```

```python
#Adding title and Labeling the Axes in the graph
# x-axis values 
x = [5, 2, 9, 4, 7] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot 
plt.scatter(x, y) 
  
# Adding Title
plt.title("GeeksFoeGeeks")
  
# Labeling the axes
plt.xlabel("Time (hr)")
plt.ylabel("Position (Km)")
  
# function to show the plot 
plt.show()
```

```python
#Multiple Graphs
x = [1, 2, 3, 4, 5] 
y = [1, 4, 9, 16, 25] 
plt.scatter(x, y) 
  
# function to show the plot 
plt.show()
  
plt.plot(x, y)
  
# function to show the plot 
plt.show()
```
